## Fine-tuning dataset: compare jsonl and hdf5 entries


```python
import glob
import json
import h5py
from transformers import AutoTokenizer, AutoModelForCausalLM
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf")
```

```python
input_dir = 'ENTER INPUT LOCATION'

sql_files = sorted(glob.glob(f'{input_dir}/splits/*.jsonl'))
hdf_files = sorted(glob.glob(f'{input_dir}/*.hdf5'))

for sql_file, hdf_file in zip(sql_files, hdf_files):

    sql_data = []
    # Read sql file
    with open(sql_file) as f:
        for line in f:
            sql_data.append(json.loads(line))
    
    hdf_data = []
    # Read hdf file
    f = h5py.File(hdf_file, 'r')
    for i in range(f['input_ids'].shape[0]):
        hdf_data.append(tokenizer.decode(f['input_ids'][i,:]))
    
    break

for d_sql, d_hdf in zip(sql_data, hdf_data):
    print(f"{d_sql['prompt']}%%%%%%%%{d_sql['completion']}")
    print('----------------------------------------------------------------')
    print(f"{d_hdf}")
```
