# SambaStudio endpoint

## Non streaming

```python

llm = SambaNovaEndpoint(
            streaming=False,
            model_kwargs={
                "do_sample": True, 
                "temperature": 0.01,
                "max_tokens_to_generate": 256,
                # "repetition_penalty": {"type": "float", "value": "1"},
                # "top_k": {"type": "int", "value": "50"},
                # "top_logprobs": {"type": "int", "value": "0"},
                # "top_p": {"type": "float", "value": "1"}
            }
        ) 
```

## Streaming

```python
llm=SambaNovaEndpoint(
    streaming=True,
    model_kwargs={
        "do_sample": True,
        "max_tokens_to_generate": 256,
        "temperature": 0.01,
        # "repetition_penalty": {"type": "float", "value": "1"},
        # "top_k": {"type": "int", "value": "50"},
        # "top_logprobs": {"type": "int", "value": "0"},
        # "top_p": {"type": "float", "value": "1"}
        }
    )
```

```python
for chunk in llm.stream("tell me a 50 word tale"):
    data = json.loads(json.loads(chunk)["data"])["stream_token"]
    print(data, end="", flush=True)
```

# Embeddings

```python
import requests
```

```python
http_session = requests.Session()
```

```python
API_BASE_PATH = "/api"
host_url = os.environ.get("EMBED_BASE_URL")
def _get_full_url(path: str) -> str:
    """
    Return the full API URL for a given path.

    :param str path: the sub-path
    :returns: the full API URL for the sub-path
    :rtype: str
    """
    return f"{host_url}{API_BASE_PATH}{path}"
```

```python
project = os.environ.get("EMBED_PROJECT_ID")
endpoint = os.environ.get("EMBED_ENDPOINT_ID")                 
key = os.environ.get("EMBED_API_KEY")

data = {"inputs": ["this is an example"]}

response =http_session.post(
    _get_full_url(f"/predict/nlp/{project}/{endpoint}"),
    headers={"key": key},
    json=data,
    )

print(response)
print(response.json())

```
